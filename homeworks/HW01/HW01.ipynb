## Создание массивов

```python
import numpy as np

# Из списка
arr1 = np.array([1, 2, 3])
print('Массив из списка:', arr1)

# С zeros
arr2 = np.zeros(3)
print('Массив нулей:', arr2)

# С ones
arr3 = np.ones(3)
print('Массив единиц:', arr3)

# С arange
arr4 = np.arange(5)
print('Последовательность:', arr4)
```

## Атрибуты

```python
import numpy as np

arr = np.array([1, 2, 3, 4])
print('Форма:', arr.shape)
print('Тип данных:', arr.dtype)
```

## Индексирование и срезы

```python
import numpy as np

arr = np.array([10, 20, 30, 40])
print('Массив:', arr)
print('Элемент [1]:', arr[1])
print('Срез [1:3]:', arr[1:3])
```

## Операции и статистика

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print('a + b:', a + b)
print('a * b:', a * b)
print('Среднее a:', a.mean())
print('Сумма a:', a.sum())
```
